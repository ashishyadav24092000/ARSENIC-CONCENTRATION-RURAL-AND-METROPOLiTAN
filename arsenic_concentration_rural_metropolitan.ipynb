# Problem Statement ::

## WE wish to determine if there is any difference in the mean arsenic concentrations between the two given communities?

# Solution::

## Method-1 ---> Statistical and algorithmic approach::

# So Final inference/insight using this mathematical approach is that ----> there is a significant evidence to conclude that mean arsenic concentration in the drinking water in rural Arizona is different from the mean arsenic concentration in metropolitan phoenix drinking water. 

# Method-2 -----> Using python as a technological tool to save our time and maximize Efficiency.

```python
## Importing all the required libraries
import math
import numpy as np
from scipy import stats
## scipiy.stats --->This module contains a large number of probability distributions,
## summary and frequency statistics, correlation functions and statistical tests, 
## masked statistics, kernel density estimation, quasi-Monte Carlo functionality, and more
```

## Step-1 :: The parameters of interest are the mean arsenic concentration for the two geographic regions, say 1 and 2 and we are interested in determining whether 1 = 2 or not.

## Step-2:  Develop the hypotheses

        Null hypothesis ----> H0 : 1 = 2
        Alternative hypothesis -----> Ha : 1  2


## Step-3: Specify the Level of Significance

          = significance level = 0.05 (say)


## Step-4: Compute the value of test statistic

## Method-1 ---> Computing t0 using function definition:

```python
metro_phoenix_community_1 = np.array([3,7,25,10,15,6,12,25,15,7])
rural_arizona_community_2 = np.array([48,44,40,38,33,21,20,12,1,18])

## Calculating mean and standard deviation for the phoenix community
x1_bar = np.mean(metro_phoenix_community_1)
print("Mean for the phoenix community is :: ",x1_bar)

std_1  = np.std(metro_phoenix_community_1)
print("Standard deviation for the phoenix community_1 is :: ",std_1)

## Calculating the mean and standard deviation of rural Arizona
x2_bar = np.mean(rural_arizona_community_2)
print("Mean for the arizona community is :: ",x2_bar)

std_2  = np.std(rural_arizona_community_2)
print("Standard deviation for the arizona community is :: ",std_2)

def test_stat(array_1,array_2,x1,x2,s1,s2):
    diff = x1-x2
    s1 = (s1**2)/len(array_1)
    s2 = (s2**2)/len(array_2)
    upon_root = (math.sqrt(s1+s2))
    return (diff/upon_root)

t0 = test_stat(metro_phoenix_community_1,rural_arizona_community_2,x1_bar,x2_bar,std_1,std_2)
print("Test statistic t0 = ", t0)
```

# --------------------------------------   OR  --------------------------------------------

## Method-2 ----> Using python libraries and their bulit-in functions

```python
stats.ttest_ind(metro_phoenix_community_1,rural_arizona_community_2,equal_var = False)
```

## So, test_statistic   ---->  t = -2.916 or -2.76

## Step-5 : Compute the Degree of Freedom(D.O.F) for t-test using python and this formula

```python
def degree_of_freedom(data_1,data_2):
    x1_bar = np.mean(data_1)
    x2_bar = np.mean(data_2)
    std_1 = np.std(data_1)
    std_2 = np.std(data_2)
    ##calculating (degree of freedom) using the formula mentioned above in algo. approach
    std_1 = ((std_1**2)/len(data_1))
    std_2 = (std_2**2)/len(data_2)
    numerator = ((std_1)+(std_2))**2
    
    denom_1 = (std_1**2)/(len(data_1)-1)
    denom_2 = (std_2**2)/(len(data_2)-1)
    denominator = denom_1 + denom_2
    
    return (numerator/denominator)

d_o_f = degree_of_freedom(metro_phoenix_community_1,rural_arizona_community_2)
print("degree of freedom =  = ",d_o_f)
```

## So, degree of freedom =  = 13 

## Step-6 : Determine the rejection rule :
    Using  = 0.05, we would reject Null hypothesis ----> H0 : 1 = 2
        1. If t0 > t(0.025,13)
        
                    OR
                    
        2. If t0 < -t(0.025,13)

## So, we have already calculated t0 = -2.916 

## Step-7 : Calculate the t-value at  /2 = 0.025, degree of freedom = 13 using pyhton

```python
t_alpha_by_2_dof = stats.t.ppf(0.025,13)
print("t-value at /2 as 0.025, degree of freedom as 13 is :: ",t_alpha_by_2_dof)
```

## Since, t0 =  -2.76    <    t(0.025,13) = -2.16  => So,we would reject the null hypothesis.

## Using the p-value approach here also the result would be the same i.e. 
            p = 0.015 <  = 0.05 => reject H0.
